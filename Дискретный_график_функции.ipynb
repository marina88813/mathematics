```python
import matplotlib.pyplot as plt

X = [1, 2, 3]
Y = [1, 4, 9]

def f(x):
    return x**2

# Все возможные пары X × Y (декартово произведение)
all_pairs = [(x, y) for x in X for y in Y]

# График функции Γ = {(x, f(x))}
graph_pairs = [(x, f(x)) for x in X]

# Визуализация
plt.figure(figsize=(6, 6))

# Рисуем все пары X×Y (серые кружочки)
for (x, y) in all_pairs:
    plt.scatter(x, y, color='lightgray', s=100, zorder=1)

# Рисуем точки графика функции (красные, крупнее)
for (x, y) in graph_pairs:
    plt.scatter(x, y, color='red', s=200, zorder=2)

# Настройки графика
plt.xticks(X, [str(x) for x in X])
plt.yticks(Y, [str(y) for y in Y])
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('График функции f(x)=x² как множество пар (x, f(x))')
plt.grid(True, linestyle='--', alpha=0.5)

# Показываем легенду (создаём заглушки)
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Все пары X×Y',
           markerfacecolor='lightgray', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='График Γ = {(x, f(x))}',
           markerfacecolor='red', markersize=12)
]
plt.legend(handles=legend_elements, loc='upper left')

plt.show()

print("X =", X)
print("Y =", Y)
print("X × Y =", all_pairs)
print("Γ =", graph_pairs)
```
